```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/Template:COVID-19_pandemic_data#covid19-container"
html=urlopen(url)
```

```python
html
```

```python
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
a=soup.find_all('a')
```

```python
tab=soup.find_all('th')
```

```python
import re
lst=[]
table = soup.find('tbody') 
for i in table.stripped_strings:
    text=repr(i)
    if re.search("[[\S]]",text):
        continue
    
    lst.append(text)
```

```python
for i in lst:
    if i=="'Ref.'":
        lst.remove(i)
    
```

```python
lst
```

```python
for n in range(len(lst)):
    word=lst[n]
    new_string = word.replace("'", "")
    lst[n]=new_string
```

```python
lst
```

```python
df = pd.DataFrame(np.array(lst).reshape(342,4))
```

```python
data=df[:50]
```

```python
data.columns = data.iloc[0]
```
